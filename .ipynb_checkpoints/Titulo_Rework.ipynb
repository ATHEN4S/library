```python
livros = livros[['Unnamed: 0', 'titulo', 'autor', 'paginas', 'editora','genero']]
livros.columns = ['index', 'titulo', 'autor', 'n_pags', 'editora', 'genero']


livros.dtypes
```

```python
livros
```

```python
def checagem_titulo(a):
  for i in range(len(a)):
    if ',' in a['titulo'][i]:
      a['titulo'][i] = a['titulo'][i].replace(",", ";" )
```

```python
checagem_titulo(livros)
```

```python
checagem_titulo(livros)
```
