```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vitorStein/livros-Skoob/master/dados.csv")
```

```python
df
```

```python
df.drop(['male','female','leram','lendo','querem_ler','abandonos','resenha','relendo','ISBN_13','idioma','descricao'], inplace=True, axis=1)
```

```python
df
```

```python
df = df.dropna(axis=0)
df = df.reset_index()
del df['index']
df
```

```python
livro = 0

for frase in df['genero']:
    for letra in frase:
        if letra == ",":
            df.drop(df.index[livro],axis = 0,inplace = True)
            break
    
    livro += 1
    df = df.reset_index()
    del df['index']
    
    if livro == 6751:
        break
```

```python
df = df.reset_index()
del df['index']
df
```

```python
livro = 0

for frase in df['titulo']:
    for letra in frase:
        if letra == ",":
            df.drop(df.index[livro],axis = 0,inplace = True)
            break
    
    
    livro += 1
    df = df.reset_index()
    del df['index']
    
    if livro == 6234:
        break
```

```python
df = df.reset_index()
del df['index']
df
```
