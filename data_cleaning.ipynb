```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vitorStein/livros-Skoob/master/dados.csv")
```

```python
df
```

```python
df.drop(['male','female','leram','lendo','querem_ler','abandonos','resenha','relendo','ISBN_13','idioma'], inplace=True, axis=1)
```

```python
df
```

```python
df = df.dropna(axis=0)
df = df.reset_index()
```

```python
del df['index']
df
```

```python
livro = 0
genero = False

for frase in df['genero']:
    for letra in frase:
        if letra == "/":
            genero = True
    
    if genero == False:
        df.drop(df.index[livro],axis = 0,inplace = True)
    
    livro += 1
    genero = False
    df = df.reset_index()
    del df['index']
```
